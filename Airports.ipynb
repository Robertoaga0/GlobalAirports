```python
import requests
from bs4 import BeautifulSoup

# URL de la pgina de Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic"

# Hacemos la solicitud HTTP
response = requests.get(url)
response.raise_for_status()  # Lanza un error si la respuesta no es 200

# Parseamos el HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Buscamos todas las tablas en la pgina
tables = soup.find_all('table', {'class': 'wikitable'})

# Extraemos la primera tabla
table = tables[0]

# Convertimos la tabla HTML a DataFrame
passengers = pd.read_html(str(table))[0]

# Mostramos las primeras filas
print(passengers.head())

# Guardamos el resultado en CSV
passengers.to_csv('passengers.csv', index=False)
```

### LIMPIEZA ###

```python
passengers.fillna(0,inplace=True)

passengers
```

```python
Limpieza_passengers = ['Total passengers', '% change']
for col in Limpieza_passengers:
    passengers[col] = passengers[col].astype(str).str.lstrip('0')

passengers
```

```python
# Cdigo para aislar el codigo IATA de la columna

passengers['IATA'] = passengers['Code (IATA/ICAO)'].str.split(r'[/ (]').str[0]

passengers
```

```python
passengers.to_csv("passengers_clean.csv")
```
